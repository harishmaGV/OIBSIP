```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming 'data' is your DataFrame with corrected column names
data.columns = [
    "Region", "Date", "Frequency",
    "Estimated Unemployment Rate(%)",
    "Estimated Employed",
    "Estimated Labour Participation Rate(%)",
    "Area"
]

# Pie Chart: Estimated Employed by Region
plt.figure(figsize=(8, 8))
employed_by_region = data.groupby("Region")["Estimated Employed"].sum()
plt.pie(
    employed_by_region,
    labels=employed_by_region.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("Set2")
)
plt.title("Estimated Employed by Region")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10, 6))
correlation = data.corr(numeric_only=True)  # Compute correlation for numeric columns
sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5
)
plt.title("Correlation Heatmap")
plt.show()
```
